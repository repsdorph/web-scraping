## Set-up of Mechanize

Select python 2.7 (For those who work in 3.x normaly

```python
#!/usr/bin/python2.7
```

Import needed liberyse. The "passwords" libery is just a way for me to hide my password

```python
import mechanize
import cookielib

import passwords
```

Initialize our Mechanize Browser instance

```python
# Browser
br = mechanize.Browser()
```

Import cookieJar to handels the cookies

```python
# Cookie Jar
cj = cookielib.LWPCookieJar()
br.set_cookiejar(cj)
```

Make a bit of magic, set the browser object to handel diffrent situations. gzip gives an warning, but that's fine.

```python
# Browser options
br.set_handle_equiv(True) # handling and response body debug printing
br.set_handle_gzip(True) # compressing content before sending it
br.set_handle_redirect(True) # Accepts redirects
br.set_handle_referer(True) # Handels refreshes of the site
br.set_handle_robots(False)  # Ignore what robots.txt says
```

If the site want to redirect us, we accept, but only if it do it in under 1 second.

```python
# Follows refresh 0 but not hangs on refresh > 0
br.set_handle_refresh(mechanize._http.HTTPRefreshProcessor(), max_time=1)
```

(If we want debug-messages)

```python
# Want debugging messages?
# br.set_debug_http(True)
# br.set_debug_redirects(True)
# br.set_debug_responses(True)
```

Tell Mechanice what headers to send. The important one is what browser we use. If it is not send, it send a "I'm a python-program"-header

```python
# User-Agent (this is cheating, ok?)
br.addheaders = [('User-agent',
                  'Mozilla/5.0 (X11; U; Linux i686; en-US; rv:1.9.0.1) Gecko/2008071615 Fedora/3.0.1-1.fc9 Firefox/3.0.1')]
```

## Starts doing stuff on the web

Opening the page we are going to login from.

```python
login_url = 'http://portalen.dtu.dk'
br.open(login_url)
```

Find the forms on the site, so we know which to log in from

```python
# Show the available forms
for f in br.forms():
    print(f)
```

We need to select the first form (form 0) to start being able to search forms. Then we fill in our login credentials and submit

```python
br.select_form(nr = 0)
br['ctl00$ContentPlaceHolder1$Textbox_Username'] = 's100322'
br['ctl00$ContentPlaceHolder1$TextBox_Password'] = passwords.dtu()
br.submit()
```

We are now loged in, but DTU is a bit wierd, so we look at what links we have

```python
# Looking at some results in link format
for l in br.links():
    print(l.text)
print("------")
for l in br.links():
    print(l)
```

We only have one link, which is a bit wierd forward who Mecanize apperentyl don't handel. No worry - we just click at the "here" link. (We could also just go to the next side, but it is more fun to click ^^)

```python
# We find a link named "here", who is to the real side with a login-token
req = br.click_link(text='here')
br.open(req)
```

Again, we look at what links there are..

```python
# Looking at some results in link format
for l in br.links():
    print(l)
```

Way to many links, but there are a way to search with regex! (Remenber that the regex is case-sensitive)

```python
# Looking at some results in link format
for l in br.links(url_regex='Kantine'):
    print(l)
```

There are only one link with "Kantine" as the text - lets click on that

```python
# Clicking on the link and opening the page
req = br.click_link(text='Kantinen')
br.open(req)
```
